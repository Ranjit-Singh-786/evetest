```python
stream = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": "Write 10 line of poem"}
    ],
    stream=True
)

for chunk in stream:
    print(chunk.choices[0].delta.content,end="")
```

```python
## image that present on your local 

import base64 

with open("demo_img.png","rb") as file:
    image_in_base64 = base64.b64encode(file.read()).decode("utf-8")

    
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "What is in this image?"},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{image_in_base64}"   # data:<file type>;base64,<encoded data>
                    }
                }
            ]
        }
    ]
)

print(response.choices[0].message.content)
```

```python
image_text_input_prompt = [
    {"role":"user",
     "content":[
         {"type":"text","text":"what you can see in this image"}, 
         {"type":"image_url",
          "image_url":{
              "url":f"data:image/jpeg;base64,{image_in_base64}"
          }} 
     ]}
]
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=image_text_input_prompt,
    max_tokens=25,
    temperature=0.5,
)

print(response.choices[0].message.content)
```

```python
image_text_prompt = """"
An Indian husband and wife sitting together on a traditional charpai (cot) in the middle of a mustard field, surrounded by blooming yellow flowers.
The husband has a well-groomed beard and mustache, with a lean build around 54kg, wearing a casual kurta-pajama. The wife is around 49kg,
wearing a colorful traditional saree with bangles, bindi, and a gentle smile. Both look happy and relaxed,
enjoying the serene countryside. The scene is set during golden hour in the evening, with soft warm sunlight casting a golden glow over the field.
The background features rural India scenery with natural greenery and a peaceful ambiance"""

result = client.images.generate(
    model="dall-e-3",
    prompt=image_text_prompt,
    n=1,
    size="1024x1024"
)
image_url = result.data[0].url
print(image_url)
```

```python
### Download the image 
import requests

response = requests.get(image_url)

with open("generated_image.png", "wb") as f:
    f.write(response.content)
    print("Your image is saved!")
```
